```python
TestData = pd.read_csv(filepathTest)
```

```python
TrainingData = pd.read_csv(filepathTrain)
```

```python
TrainingData
```

## Male = 1, Female = 0

```python
TrainingData["Sex"].replace({"male": 1, "female":0 }, inplace=True)
```

```python
TestData["Sex"].replace({"male": 1, "female":0 }, inplace=True)
```

```python
features = ['Pclass', 'Sex', 'SibSp', 'Parch']
```

```python
Xtrain = TrainingData[features]
```

```python
Xtest = TestData[features]
```

```python
Xtest
```

```python
target = TrainingData['Survived']
```

```python
Ytrain = target
```

```python
Ytrain.head()
```

```python
FirstModel = DecisionTreeRegressor(random_state = 1)
```

```python
FirstModel.fit(Xtrain,Ytrain)
```

```python
Pedictions = FirstModel.predict(Xtest)
```

```python
Pedictions
```

```python
pred_dataframe = pd.DataFrame({'Predictions': Pedictions})
```

```python
pred_dataframe
```

```python
Result = TestData.join(pred_dataframe)
```

```python
Result
```

```python
to_csv = Result.round({"Predictions":0})[['PassengerId', 'Predictions']]
```

```python
to_csv
```

```python
to_csv_int = to_csv.astype({"Predictions": int})
```

```python
to_csv_int
```

```python
to_csv_int.rename(columns={'Predictions': 'Survived'}, inplace=True)
```

```python
to_csv_int.to_csv('prediction1.csv', index=False)
```
